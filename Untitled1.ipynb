```python
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix

ground_truth_classes = ['car','speed limit','person', 'bicycle ', 'bus', 'traffic light','bike','truck','pedestrain stop','stop sign','speed limit end']
predicted_classes = ['car','speed limit','person', 'bicycle ', 'bus', 'traffic light','bike','truck','pedestrain stop','stop sign','speed limit end']

# Compute precision, recall, and F1-score for each class
precision_per_class = precision_score(ground_truth_classes, predicted_classes, average=None)
recall_per_class = recall_score(ground_truth_classes, predicted_classes, average=None)
f1_per_class = f1_score(ground_truth_classes, predicted_classes, average=None)



# Print metrics for each class
for i, car in enumerate(sorted(set(ground_truth_classes + predicted_classes))):
    print("Class:", car)
    print("  Precision:", precision_per_class[i])
    print("  Recall:", recall_per_class[i])
    print("  F1-score:", f1_per_class[i])


all_classes = sorted(set(ground_truth_classes + predicted_classes))

# Compute confusion matrix
conf_matrix = confusion_matrix(ground_truth_classes, predicted_classes, labels=all_classes)

# Loop through each class and compute TP, TN, FP, FN
for i, car in enumerate(all_classes):
    TP = conf_matrix[i, i]
    FP = conf_matrix[:, i].sum() - TP
    FN = conf_matrix[i, :].sum() - TP
    TN = conf_matrix.sum() - TP - FP - FN
    print(f"Class: {car}")
    print(f"  True Positives (TP): {TP}")
    print(f"  True Negatives (TN): {TN}")
    print(f"  False Positives (FP): {FP}")
    print(f"  False Negatives (FN): {FN}")
    print()


# Calculate overall TP, TN, FP, FN
TP = conf_matrix.diagonal().sum()
FP = conf_matrix.sum(axis=0).sum() - TP
FN = conf_matrix.sum(axis=1).sum() - TP
TN = conf_matrix.sum() - TP - FP - FN

# Print overall TP, TN, FP, FN
print(f"Overall True Positives (TP): {TP}")
print(f"Overall True Negatives (TN): {TN}")
print(f"Overall False Positives (FP): {FP}")
print(f"Overall False Negatives (FN): {FN}")




```

```python
import os

# Path to the directory containing training images
train_dir = '/content/drive/MyDrive/kaggle/Kaggal_1/train/train/images'

# Count the number of files (images) in the directory
num_train_images = len(os.listdir(train_dir))

print("Number of training images:", num_train_images)


import os

# Path to the directory containing validation images
validation_dir = '/content/drive/MyDrive/kaggle/Kaggal_1/valid/valid/images'

# Path to the directory containing test images
test_dir = '/content/drive/MyDrive/kaggle/Kaggal_1/test/test/images'

# Function to count the number of images in a directory
def count_images(directory):
    count = 0
    # Traverse through each file in the directory
    for file_name in os.listdir(directory):
        # Check if it's a file (image)
        if os.path.isfile(os.path.join(directory, file_name)):
            count += 1
    return count

# Count the number of images in the validation directory
num_validation_images = count_images(validation_dir)

# Count the number of images in the test directory
num_test_images = count_images(test_dir)

# Print the counts
print(f"Number of validation images: {num_validation_images}")
print(f"Number of test images: {num_test_images}")

```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Assuming you have true labels and predicted labels for 11 elements
true_labels = [0, 1, 1, 0, 1, 0, 1, 1, 0, 1, 0]  # True labels (ground truth)
predicted_labels = [0, 1, 0, 0, 1, 0, 1, 0, 0, 1, 0]  # Predicted labels

# Now you can compute accuracy, precision, recall, and other metrics using these lists

# Compute evaluation metrics
accuracy = accuracy_score(true_labels, predicted_labels) * 100
precision = precision_score(true_labels, predicted_labels) * 100
recall = recall_score(true_labels, predicted_labels) * 100
f1 = f1_score(true_labels, predicted_labels) * 100
conf_matrix = confusion_matrix(true_labels, predicted_labels)

# Print the computed metrics
print(f"Accuracy: {accuracy:.2f}%")
print(f"Precision: {precision:.2f}%")
print(f"Recall: {recall:.2f}%")
print(f"F1-score: {f1:.2f}%")
print(f"Confusion Matrix:\n{conf_matrix}")

```

```python
import numpy as np
from sklearn.metrics import average_precision_score

# Generate random data for 11 classes
num_samples = 1000

# True labels (ground truth) for 11 classes
true_labels = np.random.randint(0, 11, size=num_samples)

# Predicted labels for 11 classes
predicted_labels = np.random.randint(0, 11, size=num_samples)

# Ground truth bounding boxes for 11 classes
ground_truth_boxes = []
for _ in range(num_samples):
    xmin = np.random.randint(0, 100)
    ymin = np.random.randint(0, 100)
    xmax = xmin + np.random.randint(10, 50)
    ymax = ymin + np.random.randint(10, 50)
    ground_truth_boxes.append([xmin, ymin, xmax, ymax])

# Predicted bounding boxes for 11 classes
predicted_boxes = []
for _ in range(num_samples):
    xmin = np.random.randint(0, 100)
    ymin = np.random.randint(0, 100)
    xmax = xmin + np.random.randint(10, 50)
    ymax = ymin + np.random.randint(10, 50)
    predicted_boxes.append([xmin, ymin, xmax, ymax])

# Calculate Intersection over Union (IoU) for each predicted bounding box
ious = []
for predicted_box, ground_truth_box in zip(predicted_boxes, ground_truth_boxes):
    intersection = max(0, min(predicted_box[2], ground_truth_box[2]) - max(predicted_box[0], ground_truth_box[0])) * \
                   max(0, min(predicted_box[3], ground_truth_box[3]) - max(predicted_box[1], ground_truth_box[1]))
    union = (predicted_box[2] - predicted_box[0]) * (predicted_box[3] - predicted_box[1]) + \
            (ground_truth_box[2] - ground_truth_box[0]) * (ground_truth_box[3] - ground_truth_box[1]) - intersection
    iou = intersection / union
    ious.append(iou)

# Calculate Average Precision (AP) for each class
ap_per_class = []
for cls in range(11):
    # Convert class labels to binary (1 for the current class, 0 for other classes)
    binary_ground_truth = [1 if c == cls else 0 for c in true_labels]
    binary_predicted = [1 if c == cls else 0 for c in predicted_labels]
    # Compute AP
    ap = average_precision_score(binary_ground_truth, binary_predicted)
    ap_per_class.append(ap)

# Calculate Mean Average Precision (mAP)
mAP = np.mean(ap_per_class)

# Print results
print("Intersection over Union (IoU) for predicted bounding boxes:", np.mean(ious))
print("Average Precision (AP) per class:", ap_per_class)
print("Mean Average Precision (mAP):", mAP)


```
